```python
import numpy as np
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

%matplotlib notebook

img_rock_grid = mpimg.imread('./rock_grid.jpg')
img_rock = mpimg.imread('./rock.jpg')
img_large_obstacle = mpimg.imread('./large_obstacle.jpg')
img_small_obstacle = mpimg.imread('./small_obstacle.jpg')

```

```python
#def color_thresh(img, rgb_thresh_l=(30, 110, 0), rgb_thresh_u =(250,250,100)):
def color_thresh(img, rgb_thresh_l=(50, 110, 0), rgb_thresh_u =(250,250,75)):
    # Create an array of zeros same xy size as img, but single channel
    color_select = np.zeros_like(img[:,:,0])
    above_thresh = (img[:,:,0] > rgb_thresh_l[0]) & (img[:,:,0] < rgb_thresh_u[0]) & \
    (img[:,:,1] > rgb_thresh_l[1]) & (img[:,:,1] < rgb_thresh_u[1]) & \
    (img[:,:,2] > rgb_thresh_l[2]) & (img[:,:,2] < rgb_thresh_u[2])
    color_select[above_thresh] = 1
    return color_select

threshed = color_thresh(img_rock)
fig = plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(threshed, cmap='gray')
plt.show()

plt.subplot(122)
plt.imshow(img_rock)
plt.show()
```
