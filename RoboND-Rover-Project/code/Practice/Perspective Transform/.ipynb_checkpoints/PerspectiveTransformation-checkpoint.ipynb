```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2


def perspective_transform(img, src, dst):
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    return warped

%matplotlib notebook
address = './examplegrid.jpg'
img = mpimg.imread(address)

source = np.float32([[13.3064,140.855],[118.468,96.9839],[199.758,96.9839],[302.339,140.855]])
destination = np.float32([[150,155], [150,145], [160,145 ], [160,155 ]]) 

warped = perspective_transform(img,source,destination)
plt.figure(1)
plt.imshow(warped)
plt.show()

```

```python
def color_thresh(img,r=160,g=160,b=160):
    select = np.zeros_like(img[:,:,0])
    select = (img[:,:,0] > r ) & ( img[:,:,1] > g) & (img[:,:,2] > b)
    img[select] = 255
    return img
threshold = 120

colorsel = color_thresh(warped.copy(), r=threshold,g=threshold,b=threshold)
plt.figure(2)
plt.imshow(colorsel, cmap='gray')
plt.show()
```

```python
ypos, xpos = colorsel.nonzero()
plt.plot(xpos, ypos, '.')
plt.xlim(0, 320)
plt.ylim(0, 160)
#plt.show()
```
